```python
# view fasta file
multiFasta = open('dna2.fasta', 'r')
lines = multiFasta.readlines()
lines
```

```python
# the multi fasta file is in a line format
# we would like to create a dictionary where each description is a key and the sequences are its value.
Seqs = {}
# header is re-defined each time it encouters a new one, it is also added as a new key, and used in collecting all related lines of sequences
header = ''
countOfSeqs = 0
for line in lines:
    line = line.replace('\n', '')
    if line.startswith('>'):
        header = line
        countOfSeqs += 1
        Seqs.update({line: ''})
        
    else:
        Seqs[header] = (list(Seqs.values())[-1] + line)
        
countOfSeqs
```

```python
# print out the length of each sequence
seqLengths = {}
for sequences, lengthsOfSeqs in zip(Seqs.keys(), Seqs.values()):
    seqLengths.update({sequences: len(lengthsOfSeqs)})

seqLengths
```

```python
# since we need both leading and lagging strands for ORFs, the reverse complement of each sequence would also be generated.
# function for calculating reverse complement
def reverseComplement(seq):
    comp = ""
    for n in seq:
        if n == "T":
            comp += "A"
        elif n == 'A':
            comp += "T"
        elif n == "C":
            comp += "G"
        elif n == "G":
            comp += "C"
        else:
            comp += "N"
  
    revcomp = ""
    for i in range(len(comp)-1,-1,-1):
        revcomp += comp[i]
    
    return(revcomp) 
```

```python
# we will create a reverse complement for each sequence and add it to the key of each identifiers
# multiple values for a key in the dictionary
# fnrSeqs denotes forwards and reverse sequences
fnrSeqs = {}
for sequences, forwardSeqs in zip(Seqs.keys(), Seqs.values()):
    reverseSeqs = reverseComplement(forwardSeqs)
    fnrSeqs.update({sequences: [forwardSeqs, reverseSeqs]})
    
fnrSeqs
```

```python
# function for finding short tandem repeats: dinucleotides and trinucelotides

def repeatsInSeq(seq):
    #
    repeatsDict = {
        "dinucleotides": ["AT","TA", "CG", "GC", "AA", "TT",
        "CC", "GG", "AC", "AG", "TC", "TG", "CA", "CT", "GA", "GT"],

        "trinucleotides": ["AAA","AAT","AAC","AAG","ATG","ATC","ATA",
        "ATT","ACA","ACT","ACC","ACG","AGA","AGT","AGC","AGG","TTT",
        "TTC","TTA","TTG","TCT","TCC", "TCA","TCG","TAT","TAC","TGT", 
        "TGC","TGG","CTT","CTC","CTA","CTG","CCT", "CCC", "CCA","CCG",
        "CAT","CAC","CAA", "CAG", "CGT","CGC","CGA","CGG","GTT","GTC",
        "GTA","GTG", "GCT","GCC","GCA","GCG","GAT","GAC","GAA","GAG",
        "GGT","GGC","GGA","GGG","TAG","TAA","TGA"]
        }
    #
    import re
    #
    diNucl = {}
    for i in repeatsDict["dinucleotides"]:
        match = re.finditer(pattern=i, string=seq)
        diNucl[i] = [index.start() for index in match]
    #
    for dn, diN in zip(diNucl.values(), diNucl.keys()):
        total = 0
        count = 0
        x = 0
        dn.extend((-7, -7))
        while x < len(dn)-2:
            x += 1
            for i in dn:
                if i == dn[x]-2 and i == dn[x+1]-4:
                    count += 1   
                elif i == dn[x]-2 and i == dn[x-2]+2:
                    count += 1                
                elif i == dn[x-2]+2 and i == dn[x-3]+4:
                    count += 1
                    print(count, 'STRs ending at pos.', i+2)
                    total += count
                    count = 0
        if total > 0:
            print(diN, 'total STRs in the sequence :', total)
    #
    triNucl = {}
    for i in repeatsDict["trinucleotides"]:
        match = re.finditer(pattern=i, string=seq)
        triNucl[i] = [index.start() for index in match]
    #
    for tn, triN in zip(triNucl.values(), triNucl.keys()):
        total = 0
        count = 0
        x = 0
        tn.extend((-7, -7))
        while x < len(tn)-2:
            x += 1
            for i in tn:
                if i == tn[x]-3 and i == tn[x+1]-6:
                    count += 1 
                elif i == tn[x]-3 and i == tn[x-2]+3:
                    count += 1
                elif i == tn[x-2]+3 and i == tn[x-3]+6:
                    count += 1
                    print(count, 'STRs ending at pos.', i+3)
                    total += count
                    count = 0
        if total > 0:
            print(triN, 'total STRs in the sequence :', total)
    #
```
